## Agentes pessoais
Programação baseada em modelos é utilizada para executar tarefas específicas.

![agente-pessoal-arquitetura.png.png](imagens/fluxo-agentes-pessoais.png)

Existem 5 níveis de agentes pessoais: L1, L2, ..., L5.

- **L1**: Executa passos simples a partir das instruções do usuário.
- **L2**: Determina de forma automática o que utilizar para excutar tarefas simples.
- **L3**: Automatiza tarefas estratégicas. Determina de forma automática quais ferramentas e quais recursos - disponiveis para executar as instruções do usuário. 
- **L4**: O agente detecta e memoriza o contexto. Ótimo para modelos de recomendação.
- **L5**: Pesonagem digital. Representa um usuario em tarefas mais complexas, como ler e interpretar e-mails, responder perguntas e até mesmo participar de reuniões de trabalhos.

> O GPT-4o como modelo performatico para agentes pessoais.

#### Planejamento

No planejamento o agente decomponhe as tarefas através do chain of thought (CoT) utilizando arvores de pensamentos. 
Também pode ser instruído atraves de prompts e também pode trabalhar com planejamento externo. Neste caso as instruções sao convertidas para PDDL, interpretadas e posteriormente volta para linguagem natural como output.
A etapda de ReAct executa uma interpretação da entrada, executa as instruções e busca entender e iteragir com as saídas geradas por ele mesmo, afim de encontrar melhores informações.

> PDDL e LLM são combinados para melhorar o raciocínio formal de modelos de linguagem grandes (LLMs). PDDL (Planning Domain Definition Language) é uma linguagem formal para planejamento automatizado, enquanto LLMs geram texto e raciocínio. O uso de PDDL permite que os LLMs executem tarefas de planejamento com maior precisão e validade lógica, por meio de técnicas como ajuste de instruções e feedback de planejadores simbólicos externos.

#### Ferramentas

Analisa a solicitação e depois considera qual tipo de tarefa e requisitos detalhados para quela tarefa. após isso seleciona a ferramenta/API mais adequada para execução.
    - Depende das instruções prescritas nas proprias ferramentas.

1. MRKL - O modelo direciona as perguntas para algums modulos especializados.
2. TALM e Toolformer - Treinados para aprender e utilizar ferramentas externas com base em notações.
3. HuggingFace - Seleciona o modelo mais adequado a ser utilizado como ferramentas. Consuta a propria base do HuggingFace para sugerir.
4. extra. MCP - Capaz de executar ferramentas especificas.

#### Memórias

A memoria é baseada na capacidade humana. A "memoria sensorial" da máquina diz respeito ao processamento inicial dos dados brutos. Como uma imagen, som e texto. 
Já no curto prazo, dados temporários são armazenados e apagados apos sua utilização.
Memoria a longo prazo é armazenadas para utlização de forma cotidiana.
A memoria explicita seriam os dados estruturados que a máquina pode consultar.
